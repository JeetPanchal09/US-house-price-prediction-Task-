### The Economy

### When the economy is going strong and unemployment goes down, housing prices tend to rise. In a strong economy, people feel more secure in their jobs and their ability to take on mortgage debt. They're also more likely to get approved for that mortgage.

### When the economy falters, however, more people find themselves laid off, having to live on unemployment compensation or accept much lower paying jobs. When this happens, fewer people are able to purchase houses, competition for them goes away and motivated sellers must lower their asking price to get a sale.

# Question :  Which factors affect the price of house the most?

```python
x_col = ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Bedrooms', 'Avg. Area Number of Rooms', 'Area Population']
y_col = ['Price']

sns.pairplot(df, 
            x_vars= x_col,
            y_vars= y_col)
plt.show()
```

```python
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Instantiate LabelEncoder
le=LabelEncoder()

# Iterate over all the values of each column and extract their dtypes
for col in df.columns:
    # Compare if the dtype is object
    if df[col].dtypes=='object':
    # Use LabelEncoder to do the numeric transformation
        df[col]=le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

X = df.drop(['Price', 'Population group', 'Address'], axis=1)
y = df['Price']

pipeline = Pipeline([
    ('std_scalar', StandardScaler())])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression(normalize=True)
lin_reg.fit(X_train,y_train)
```

```python
print("Train score:")
print(lin_reg.score(X_train, y_train))

print("Test score:")
print(lin_reg.score(X_test, y_test))
```

```python
coeff_df = pd.DataFrame(lin_reg.coef_, X.columns, columns=['Coefficient'])
coeff_df.sort_values(by='Coefficient', ascending=False)
```

```python
names = df.drop(['Price', 'Population group', 'Address'], axis=1).columns
lr_coef = lin_reg.fit(X_train, y_train).coef_
_ = plt.plot(range(len(names)), lr_coef)
_ = plt.xticks(range(len(names)), names, rotation=90)
_ = plt.ylabel('Coefficients')
plt.show()
```

### Answer : The model suggests that area income, house age and area population afftects "Price" of houses the most.
